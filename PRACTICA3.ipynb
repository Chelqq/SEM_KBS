```python
import numpy as np
import matplotlib.pyplot as plt

# Datos de entrada
X = np.array([1, 2, 3, 4])
Y = np.array([1, 2, 3, 4])

#Parmetros
w = 0.0
b = 0.0
alpha = 0.01 
epochs = 1000

n = len(X)  # Num de datos

# Pesos y sesgos para visualizar
history_w = []
history_b = []
history_mse = []

for epoch in range(epochs):
    # Prediccin del modelo
    Y_pred = w * X + b

    # Calculo del gradiente
    dw = (-2/n) * np.sum(X * (Y - Y_pred))
    db = (-2/n) * np.sum(Y - Y_pred)

    # Actualizacin de parmetros
    w = w - alpha * dw
    b = b - alpha * db

    # Clculo de MSE
    mse = np.mean((Y - Y_pred)**2)

    # Almacenar valores para la grfica
    history_w.append(w)
    history_b.append(b)
    history_mse.append(mse)

    if epoch % 100 == 0:
        print(f'Epoch {epoch}: MSE = {mse}')

print(f'Modelo final: w = {w}, b = {b}')

# Grfica de la lnea de regresin final
plt.figure(figsize=(10, 6))
plt.scatter(X, Y, color='blue', label='Datos reales')

# Predicciones con los parmetros finales
Y_pred_final = w * X + b
plt.plot(X, Y_pred_final, color='red', label='Lnea de regresin')

plt.title('Regresin Lineal Simple')
plt.xlabel('X')
plt.ylabel('Y')
plt.legend()
plt.grid(True)
plt.show()

#MSE durante el entrenamiento
plt.figure(figsize=(10, 6))
plt.plot(history_mse, color='purple', label='MSE')
plt.title('Evolucin del Error Cuadrtico Medio (MSE)')
plt.xlabel('pocas')
plt.ylabel('MSE')
plt.legend()
plt.grid(True)
plt.show()
```
